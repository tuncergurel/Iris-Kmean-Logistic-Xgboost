```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('iris.txt', header = None)
```

```python
iris.columns = ['a','b','c','d','e']
```

```python
iris
```

```python
iris_test = iris.pop('e')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
iris_test = le.fit_transform(iris_test)
```

```python
iris_test
```

```python
iris2 = iris.values
```

```python
iris2
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
```

```python
km.fit(iris2)
```

```python
km.labels_
```

```python
result = km.labels_ == iris_test
```

```python
for i in range(3):
    print('{}.cluster : '.format(i+1))
    print(iris2[km.labels_ == i])
    print('------------------------------------')
    
```
